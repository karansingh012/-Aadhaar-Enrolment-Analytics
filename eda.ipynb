# Distribution Shape Analysis

<br> Skewness measures the asymmetry of a data distribution, while kurtosis measures the tailedness (presence of outliers) in the distribution.
	<br> 	Skewness = 0  Distribution is perfectly symmetric
	<br> 	Skewness > 0  Distribution is right-skewed (long right tail)
	<br> 	Skewness < 0  Distribution is left-skewed (long left tail)
	<br> 	Kurtosis = 0  Distribution is normal (mesokurtic)
	<br> 	Kurtosis > 0  Distribution has heavy tails (more extreme outliers)
	<br> 	Kurtosis < 0  Distribution has light tails (fewer outliers)

```python
from scipy.stats import skew, kurtosis

skew(df['total_enrollment']), kurtosis(df['total_enrollment'])
```

**Our data is highly non-normal, with most values clustered at the lower end and a few extreme high values dominating the distribution.**
